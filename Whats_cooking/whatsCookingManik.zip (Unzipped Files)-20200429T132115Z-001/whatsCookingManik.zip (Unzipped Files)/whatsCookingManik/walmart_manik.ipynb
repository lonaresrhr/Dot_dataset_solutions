```python
import pandas as pd
import numpy as np
#from matplotlib import pyplot as plt
#from matplotlib.gridspec import GridSpec
#import seaborn as sns
from scipy import stats
#from scipy.special import boxcox1p

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
features = pd.read_csv('features.csv')
train = pd.read_csv('train.csv')
stores = pd.read_csv('stores.csv')
test = pd.read_csv('test.csv')
```

```python
#feat_sto = features.merge(stores, how='inner', on='Store')
feat_sto=pd.merge(train,features,on=['Store','Date','IsHoliday'],how='inner')
feat_sto=pd.merge(feat_sto,stores,on='Store',how='inner')
```

```python
feat_sto.info()
```

```python
feat_sto['Date']=pd.to_datetime(feat_sto['Date'])
feat_sto['year']=feat_sto['Date'].dt.year
feat_sto['month']=feat_sto['Date'].dt.month
feat_sto['day']=feat_sto['Date'].dt.day
del feat_sto['Date']
```

```python
from sklearn.preprocessing import LabelEncoder

feat_sto['Type']=LabelEncoder().fit_transform(feat_sto['Type'])
feat_sto['IsHoliday']=LabelEncoder().fit_transform(feat_sto['IsHoliday'])
feat_sto.dtypes
```

```python
feat_sto
```

```python
TrainData = feat_sto[feat_sto.columns.difference(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'])]
```

```python
TrainData
```

```python
TrainData.isna().mean()*100
```

```python
from scipy.stats import zscore 
from sklearn.model_selection import train_test_split
#import matplotlib.pyplot as plt

X = TrainData.drop('Weekly_Sales',axis=1)
y = TrainData['Weekly_Sales']
```

```python
# Now compute the relative Z-score of the input data, relative to the sample mean and standard deviation.
X_scaled = X.apply(zscore)

X_trainS,X_testS,y_trainS,y_testS = train_test_split(X_scaled,y,test_size=.3)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3)

from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import r2_score, mean_squared_error

dt = DecisionTreeRegressor()

dt.fit(X_trainS,y_trainS)
y_pred = dt.predict(X_testS)
print("r2 scaled DT = ",r2_score(y_testS,y_pred))
print("rmse scaled DT = ",np.sqrt(mean_squared_error(y_testS,y_pred)))

dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print("r2 Notscaled DT = ",r2_score(y_test,y_pred))
print("rmse scaled DT = ",np.sqrt(mean_squared_error(y_testS,y_pred)))
```

```python
rf = RandomForestRegressor()

rf.fit(X_trainS,y_trainS)
y_pred = rf.predict(X_testS)
print("r2 Scaled RF = ",r2_score(y_testS,y_pred))
print("rmse scaled RF = ",np.sqrt(mean_squared_error(y_testS,y_pred)))

rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print("r2 Notscaled RF = ",r2_score(y_test,y_pred))
print("rmse scaled RF = ",np.sqrt(mean_squared_error(y_testS,y_pred)))
```

```python
lr = LinearRegression()

lr.fit(X_trainS,y_trainS)
y_pred = lr.predict(X_testS)
print("r2 Scaled LR = ",r2_score(y_testS,y_pred))
print("rmse scaled LR = ",np.sqrt(mean_squared_error(y_testS,y_pred)))

lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print("r2 Notscaled LR = ",r2_score(y_test,y_pred))
print("rmse scaled LR = ",np.sqrt(mean_squared_error(y_testS,y_pred)))
```

```python
pf = PolynomialFeatures(degree=2)
polynomial_features_scaled = pf.fit_transform(X_scaled)
polynomial_features = pf.fit_transform(X)

X_trainS,X_testS,y_trainS,y_testS = train_test_split(polynomial_features_scaled,y,test_size=.3)
X_train,X_test,y_train,y_test = train_test_split(polynomial_features,y,test_size=.3)

lr.fit(X_trainS,y_trainS)
y_pred = lr.predict(X_testS)
print("r2 Scaled LR = ",r2_score(y_testS,y_pred))
print("rmse scaled LR = ",np.sqrt(mean_squared_error(y_testS,y_pred)))

lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print("r2 Notscaled LR = ",r2_score(y_test,y_pred))
print("rmse scaled LR = ",np.sqrt(mean_squared_error(y_testS,y_pred)))
```
