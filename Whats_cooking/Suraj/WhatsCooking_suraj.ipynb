```python

from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import nltk
import re
from nltk.stem import WordNetLemmatizer
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import sklearn.metrics
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
print("library Import done")
```

```python
import json

traindf = pd.read_json("train.json")

testdf = pd.read_json("test.json")

print("file Import done")
testdf.info
```

Refer [WordNetLemmatizer](https://www.machinelearningplus.com/nlp/lemmatization-examples-python/) link. Lemmatization is the process of converting a word to its base or root form. For eg, word 'creating' is converted to word 'create'.

```python
traindf['ingredients_clean_string'] = [' , '.join(z).strip() for z in traindf['ingredients']]  
traindf['ingredients_string'] = [' '.join([ WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]', ' ', line)) for line in lists]).strip() for lists in traindf['ingredients']]       

testdf['ingredients_clean_string'] = [' , '.join(z).strip() for z in testdf['ingredients']]
testdf['ingredients_string'] = [' '.join([WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]', ' ', line)) for line in lists]).strip() for lists in testdf['ingredients']]       

traindf
```

Refer [TfidfVectorizer](https://kavita-ganesan.com/tfidftransformer-tfidfvectorizer-usage-differences/#.XqMCSsgzZPY) link. TfidfVectorizer function from scikit-learn computes the word counts, IDF (Inverse Document Frequency) values, and Tf-idf scores using the same dataset. Basically like word2vec concept it maps words to numbers for training.


```python
# Data formatting, in particular vectoring texual data
##################################################
#  training data formatiing
corpustr = traindf['ingredients_string']
vectorizertr=TfidfVectorizer(use_idf=True)
tfidftr=vectorizertr.fit_transform(corpustr)
##################################################
#  testing data formatiing
corpusts = testdf['ingredients_string']
vectorizerts = TfidfVectorizer(use_idf=True)
tfidfts=vectorizertr.transform(corpusts)
##################################################

targets_tr = traindf['cuisine']

print("print corpustr")
corpustr
```

Using classification model

```python

classifier = LogisticRegression()

classifier=classifier.fit(tfidftr,targets_tr)

predictions=classifier.predict(tfidfts)
print("predictions done")

```

Make submission to csv file

```python

testdf['cuisine'] = predictions
testdf[['id'  , 'cuisine' ]].to_csv("result_suraj_cooking.csv")
print("file submission done..")
```
